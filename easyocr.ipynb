```python
import easyocr as ocr
import cv2
import numpy as np
```

```python
ocr_motoru = ocr.Reader(['en','tr'])
```

```python
with open('resim_fis.jpg','rb') as f:
  resim = f.read()
resim_ham = np.fromstring(resim, np.uint8)
resim_cv2 = cv2.imdecode(resim_ham,cv2.IMREAD_COLOR)
```

```python
from google.colab.patches import cv2_imshow

cv2_imshow(resim_cv2)
```

```python
yazilar = ocr_motoru.readtext(resim)
```

```python
for yazi in yazilar:
  print(yazi)
```

```python
for yazi in yazilar:
  cv2.rectangle(resim_cv2,[(yazi[0][0][0]),(yazi[0][0][1])], [(yazi[0][2][0]), (yazi[0][2][1])], (0,0,255),2)
```
